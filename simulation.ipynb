```python
from cvxopt import matrix, solvers
solvers.options['show_progress'] = False
```

```python
def EfficientFrontier(NumPortf, Covariance, ExpectedValues):
    NumAssets = Covariance.shape[1]
    # determine return of minimum-risk portfolio
    FirstDegree = np.zeros([NumAssets, 1])
    SecondDegree = Covariance
    Aeq = np.ones([1, NumAssets])
    beq = np.array([[1.]])
    A = -np.eye(NumAssets) # no-short constraint
    b = np.zeros([NumAssets, 1]) # no-short constriant
    x0 = 1 / NumAssets * np.ones([NumAssets,1])
    MinVol_Weights = solvers.qp(matrix(SecondDegree), matrix(FirstDegree), matrix(A), matrix(b), matrix(Aeq), matrix(beq))['x']
    MinVol_Weights = np.array(MinVol_Weights)
    MinVol_Return = np.dot(MinVol_Weights.T, ExpectedValues)[0]
    #############################################
    # determine return of maximum-return portfolio
    MaxRet_Return = max(ExpectedValues)
    # slice efficient frontier in Numportf equally thick horizontal sectors in the upper branch only
    Step = (MaxRet_Return - MinVol_Return) / (NumPortf-1)
    TargetReturns = np.arange(MinVol_Return,MaxRet_Return+0.0001, Step)

    ############################################
    # compute the Numportf compositions and risk-return coordinates of the optimal allocations relative to each slice

    # start with min vol portfolio
    Composition = MinVol_Weights.T
    Volatility = np.sqrt(np.dot(np.dot(MinVol_Weights.T, Covariance), MinVol_Weights))[0][0]
    ExpectedValue = np.dot(MinVol_Weights.T, ExpectedValues)[0]
    all_Weights = []
    all_ExpectedValue = []
    all_Volatility = []
    all_Weights.append(Composition.T)
    all_ExpectedValue.append(ExpectedValue)
    all_Volatility.append(Volatility)
    for i in range(1,NumPortf-1):
        # determine least risky portfolio for given expected return
        AEq = np.concatenate([np.ones([1, NumAssets]), ExpectedValues.reshape(-1,1).T], axis=0)
        bEq = np.array([[1],[TargetReturns[i]]])
        Weights = solvers.qp(matrix(SecondDegree), matrix(FirstDegree), matrix(A), matrix(b), matrix(AEq), matrix(bEq))['x']
        Weights = np.array(Weights)
        all_Weights.append(Weights)
        all_ExpectedValue.append(np.dot(Weights.T, ExpectedValues)[0])
        all_Volatility.append(np.sqrt(np.dot(np.dot(Weights.T, Covariance), Weights))[0][0])
    return all_ExpectedValue, all_Volatility, all_Weights
```

### bayesian robusr allocation

```python
# input
J = 1000 # number of simulations
T = 52 # number of obervations in time series
N = 20 # number of asset in the market
r = 0.4 # overall correlation
min_s = 0.1 # min volatility
max_s = 0.4 # max volatility
NumPortf = 10
p_m = 0.1 # aversion to estimation risk for mu
p_s = 0.1 # aversion to estimation risk for sigma
```

```python
# true market parameters
C = (1 - r) * np.eye(N) + r * np.ones((N,N))
step_s = (max_s - min_s) / (N-1)
s = np.arange(min_s, max_s+0.001, step_s)
S = np.dot(np.dot(np.diag(s), C), np.diag(s))
M = 2.5 * np.dot(S, np.ones((N,1))) / N
```

```python
from numpy.random import multivariate_normal
from scipy.stats.distributions import chi2
```

```python
all_all_erb = []
all_all_vrb = []
all_all_e = []
all_all_v = []
all_all_e1 = []
all_all_v1 = []
```

```python
for j in range(J):
    print(j) 
    Rets = multivariate_normal(M.flatten(), S, T)
    # sample estimate
    m_hat = np.mean(Rets, axis=0)
    S_hat = np.cov(Rets.T)
    de_hat, ds_hat, w_hat = EfficientFrontier(NumPortf, S_hat, m_hat)
    # Bayesian prior
    S0 = np.diag(np.diag(S_hat))
    m0 = 0.5 * np.dot(S0 , np.ones([N,1])) / N
    T = Rets.shape[0]
    T0 = 2 * T
    nu0 = 2 * T
    # Bayesian posterior parameters
    m_hat = m_hat.reshape(-1,1)
    T1 = T0 + T
    m1 = 1 / T1 * (m_hat * T + m0*T0)
    m1 = m1.flatten()
    nu1 = T + nu0
    S1 = 1/nu1 * (S_hat * T + S0 * nu0 + np.dot((m_hat-m0),(m_hat-m0).T)/(1/T+1/T0))
    de, ds, w1 = EfficientFrontier(NumPortf, S1, m1)
    # robustness parameters
    q_m2 = chi2.ppf(p_m, df=N)
    g_m = np.sqrt(q_m2/T1*nu1/(nu1-2))
    q_s2 = chi2.ppf(p_s, N*(N+1)/2)
    PickVol = round(0.8 * NumPortf)
    v = (ds_hat[PickVol]) ** 2
    g_s = v / (nu1 / (nu1+N+1)+np.sqrt(2*nu1*nu1*q_s2/((nu1+N+1)**3)))
    all_e = []
    all_v = []
    all_e1 = []
    all_v1 = []
    Target = []
    for k in range(NumPortf-1):
#         print(k)
        # sample allocation
        wu = w_hat[k]
        tmp_e = np.dot(M.T, wu)[0][0]
        tmp_v = np.dot(np.dot(wu.T, S), wu)[0][0]
        all_e.append(tmp_e)
        all_v.append(tmp_v)
        # bayesian allocation
        wu = w1[k]
        tmp_e1 = np.dot(M.T, wu)[0][0]
        tmp_v1 = np.dot(np.dot(wu.T, S), wu)[0][0]
        all_e1.append(tmp_e1)
        all_v1.append(tmp_v1)
        # robustness
        NewTarget = - 10**10
        ttmp_v1 = np.dot(np.dot(wu.T, S1), wu)[0][0]
        if ttmp_v1 <= g_s:
            NewTarget = np.dot(m1.reshape(-1,1).T, wu) - g_m * np.sqrt(ttmp_v1)
            NewTarget = NewTarget[0][0]
        Target.append(NewTarget)
    best_k =np.argmax(Target)
    wu = w1[best_k]
    tmp_erb = np.dot(M.T, wu)[0][0]
    tmp_vrb = np.dot(np.dot(wu.T, S), wu)[0][0]
    all_all_erb.append(tmp_erb)
    all_all_vrb.append(tmp_vrb)
    all_all_e.append(all_e)
    all_all_v.append(all_v)
    all_all_e1.append(all_e1)
    all_all_v1.append(all_v1)
```
